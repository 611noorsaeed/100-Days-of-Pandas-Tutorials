```python
import pandas as pd
```

```python
df = pd.read_csv("dummy.cvs")
```

```python
df
```

# Sorting by a single column (e.g., Age)

```python
df.sort_values(by="Age")
```

# Sorting by multiple columns (e.g., Department and Salary)

```python
df.sort_values(by=['Department','Salary'])
```

# Sorting by index

```python
df.sort_index()
```

# Sorting in descending order

```python
df.sort_values(by='Age',ascending=False)
```

# Sorting and keeping the original index

```python
df.sort_values(by='Age').reset_index(drop=True)
```

# Sorting and creating a new index


```python
df.sort_values(by='Salary').reset_index(drop=True)
```

```python
df
```

# Sorting by a custom order (e.g., Department)


```python
custom_sort_order = ['IT', 'HR', 'Marketing', 'Finance']
df[df['Department'].isin(custom_sort_order)]
```

# Sorting by a custom function (e.g., sorting by name length)

```python
df.reindex(df['Name'].str.len().sort_values().index)
```

# Sorting by a custom function and ignoring NaN values
df

```python
df.sort_values(by='Age',key=lambda x:x.fillna(0))
```
