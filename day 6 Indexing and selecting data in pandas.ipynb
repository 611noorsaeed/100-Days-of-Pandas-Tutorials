# Accessing data at different levels of a MultiIndex

```python
# Creating a DataFrame with MultiIndex columns
data = {
    ('Group 1', 'A'): [1, 2, 3],
    ('Group 1', 'B'): [4, 5, 6],
    ('Group 2', 'C'): [7, 8, 9],
    ('Group 2', 'D'): [10, 11, 12]
}
df = pd.DataFrame(data)
df.columns.names = ['Group', 'Subgroup']
df
```

```python
df['Group 1']
```

```python
df['Group 2']
```

```python
df['Group 1']['A']
```

```python
df.loc[:,('Group 1', 'B')]
```

```python
df.loc[:, ('Group 2', 'C'):('Group 2', 'D')]
```
